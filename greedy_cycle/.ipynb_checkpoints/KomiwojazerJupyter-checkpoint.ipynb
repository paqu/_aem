```python
def podzialInt (linia):
    tab2=np.array                   #miejsce na linie na tabelke
    tab=[]
    #print(linia)
    for i in linia.split(" "):      #podzial
       try:
           tab.append(int(i))
       except ValueError:
            try:
                tab.append(float(i))
            except ValueError:
                return False
    tab2=tab

    return tab2
```

```python
import tsplib95
import numpy as np
import matplotlib.pyplot as plt 
import random as rd
#Wczytywanie instancji kroa100 i krob100 (w jednym z formatów w jakim są dostępne) i obliczanie macierzy odległości.
problem=tsplib95.load_problem("instances/kroA100.tsp")
lista=list(problem.get_edges())
#problem.wfunc(1,2)
tab=[]
for x,y in lista:
    tab.append(problem.wfunc(x,y))
tab2=np.array(tab)
odleglosci=tab2.reshape(x,y)
print(odleglosci)
########################################################################################
#od teraz jest macierz odleglosci jako odleglosci. Wszystko leci od niej
#Określenie ilości miast i miasta startowego
iloscMiast=odleglosci.shape
iloscMiast=iloscMiast[0]
#startMiasto=rd.randint(0,iloscMiast-1)
startMiasto=61
odleglosciBackup=np.array(odleglosci)
odleglosci[:,startMiasto]=0
#print(odleglosci)
#print(startMiasto)
sciezka=[]
sciezka.append(startMiasto)
kosztLaczny=0

#Znajdowanie najbliższego sąsiada dla pierwszego miasta
najblizszyOdleglosc=min(min(odleglosci[startMiasto,np.where(odleglosci[startMiasto,:]>0)]))
najblizszyNumer=0
najblizszaOdleglosc=0
for i,v in enumerate(odleglosci[startMiasto,:]):
    if najblizszaOdleglosc==0 and v>0:
        najblizszaOdleglosc=v
    if najblizszaOdleglosc>v and v>0:
        najblizszaOdleglosc=v
        najblizszyNumer=i
sciezka.append(najblizszyNumer)
odleglosci[:,najblizszyNumer]=0
kosztLaczny+=najblizszyOdleglosc
#print(sciezka)
#print(odleglosci)

#Znajdowanie kolejnego

for i in range (round((iloscMiast)/2)-2):#dla punktów
#for i in range (1):#dla punktów
    bestDnKoszt=-10
    bestDn=-1
    insertDn=-1
    for punkt in range (len(sciezka)):      #dla boków
        d1=sciezka[punkt]
        d0=sciezka[punkt-1]
        kosztD0D1=odleglosciBackup[d0,d1]
        kosztD0Dn=-1
        kosztMinimalny=-10
        kolejny=-1
        #print(kosztD1Di)
        for dn,kosztD0Dn in enumerate(odleglosci[sciezka[punkt-1],:]):
            if(kosztD0Dn!=0):
                kosztD1Dn=odleglosciBackup[d1,dn]
                kosztDn=kosztD0Dn+kosztD1Dn-kosztD0D1
                if(kosztMinimalny==-10) or kosztMinimalny>kosztDn:
                    kolejny=dn
                    kosztMinimalny=kosztDn
        if bestDnKoszt==-10 or kosztMinimalny<bestDnKoszt:
            bestDnKoszt=kosztMinimalny
            bestDn=kolejny
            insertDn=punkt
    sciezka.insert(insertDn+1,bestDn)
    odleglosci[:,sciezka[insertDn+1]]=0

print(sciezka)
print(odleglosci)

kosztLaczny=odleglosciBackup[sciezka[0],sciezka[-1]]
for i in range (len(sciezka)-1):
    kosztLaczny+=odleglosciBackup[sciezka[1],sciezka[i+1]]
print(kosztLaczny)
#sciezka2=np.array(sciezka)
#print(sciezka2.shape)
```
